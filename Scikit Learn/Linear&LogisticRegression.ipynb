# Scikit learn: Linear Regression & Logistic Regression

```python
# Import necessary libraries
import numpy as np
from sklearn.datasets import load_diabetes, load_breast_cancer
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Linear Regression Model

```python
# 1. Linear Regression Model
# Load diabetes dataset (regression problem)
diabetes = load_diabetes()
X_lin, y_lin = diabetes.data, diabetes.target

# Standardize features
scaler_lin = StandardScaler()
X_lin_scaled = scaler_lin.fit_transform(X_lin)

# Split the data
X_lin_train, X_lin_test, y_lin_train, y_lin_test = train_test_split(
    X_lin_scaled, y_lin, test_size=0.2, random_state=42
)

# Create and train Linear Regression model
linear_model = LinearRegression()
linear_model.fit(X_lin_train, y_lin_train)

# Make predictions and evaluate
lin_predictions = linear_model.predict(X_lin_test)
mse = mean_squared_error(y_lin_test, lin_predictions)
r2 = r2_score(y_lin_test, lin_predictions)

print("Linear Regression Results:")
print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python
# 1. LINEAR REGRESSION ANALYSIS
diabetes = load_diabetes()
X_lin, y_lin = diabetes.data, diabetes.target

# Create a DataFrame for better analysis
df_diabetes = pd.DataFrame(X_lin, columns=diabetes.feature_names)
df_diabetes['target'] = y_lin

# Statistical Analysis
print("LINEAR REGRESSION DATASET ANALYSIS:")
print("\nDataset Shape:", df_diabetes.shape)
print("\nFeature Statistics:")
print(df_diabetes.describe())

# Correlation Analysis
plt.figure(figsize=(10, 8))
sns.heatmap(df_diabetes.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix - Diabetes Dataset')
plt.tight_layout()
plt.show()

# Feature Importance Analysis
scaler_lin = StandardScaler()
X_lin_scaled = scaler_lin.fit_transform(X_lin)
X_lin_train, X_lin_test, y_lin_train, y_lin_test = train_test_split(
    X_lin_scaled, y_lin, test_size=0.2, random_state=42
)

linear_model = LinearRegression()
linear_model.fit(X_lin_train, y_lin_train)

# Feature importance based on coefficients
feature_importance_lin = pd.DataFrame({
    'feature': diabetes.feature_names,
    'importance': abs(linear_model.coef_)
})
feature_importance_lin = feature_importance_lin.sort_values('importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='importance', y='feature', data=feature_importance_lin)
plt.title('Feature Importance in Linear Regression')
plt.show()

# Cross-validation scores
cv_scores_lin = cross_val_score(linear_model, X_lin_scaled, y_lin, cv=5, scoring='r2')
print("\nCross-validation R² scores:", cv_scores_lin)
print("Mean R² score:", cv_scores_lin.mean())
print("Standard deviation:", cv_scores_lin.std())
```

## Logistic Regression Model

```python
# 2. Logistic Regression Model
# Load breast cancer dataset (classification problem)
cancer = load_breast_cancer()
X_log, y_log = cancer.data, cancer.target

# Standardize features
scaler_log = StandardScaler()
X_log_scaled = scaler_log.fit_transform(X_log)

# Split the data
X_log_train, X_log_test, y_log_train, y_log_test = train_test_split(
    X_log_scaled, y_log, test_size=0.2, random_state=42
)

# Create and train Logistic Regression model
logistic_model = LogisticRegression(
    random_state=42,
    max_iter=2000,      # Increased iterations for convergence
    tol=1e-4,          # Tolerance for stopping criteria
    C=1.0,             # Inverse of regularization strength
    solver='lbfgs'     # Optimizer
)
logistic_model.fit(X_log_train, y_log_train)

# Make predictions and evaluate
log_predictions = logistic_model.predict(X_log_test)
accuracy = accuracy_score(y_log_test, log_predictions)

print("\nLogistic Regression Results:")
print(f"Accuracy: {accuracy:.2f}")
print("\nClassification Report:")
print(classification_report(y_log_test, log_predictions))

# Visualize results (for the first feature of each dataset)
plt.figure(figsize=(12, 5))

# Linear Regression Plot
plt.subplot(1, 2, 1)
plt.scatter(X_lin_test[:, 0], y_lin_test, color='blue', label='Actual')
plt.scatter(X_lin_test[:, 0], lin_predictions, color='red', label='Predicted')
plt.title('Linear Regression: Actual vs Predicted')
plt.xlabel('Feature 1')
plt.ylabel('Target')
plt.legend()

# Logistic Regression Plot
plt.subplot(1, 2, 2)
plt.scatter(X_log_test[:, 0], y_log_test, color='blue', label='Actual')
plt.scatter(X_log_test[:, 0], log_predictions, color='red', label='Predicted')
plt.title('Logistic Regression: Actual vs Predicted')
plt.xlabel('Feature 1')
plt.ylabel('Target')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 2. LOGISTIC REGRESSION ANALYSIS
cancer = load_breast_cancer()
X_log, y_log = cancer.data, cancer.target

# Create DataFrame for analysis
df_cancer = pd.DataFrame(X_log, columns=cancer.feature_names)
df_cancer['target'] = y_log

print("\nLOGISTIC REGRESSION DATASET ANALYSIS:")
print("\nDataset Shape:", df_cancer.shape)
print("\nClass Distribution:")
print(df_cancer['target'].value_counts(normalize=True))

# Correlation Analysis for top features
plt.figure(figsize=(12, 10))
top_features = df_cancer.corr()['target'].sort_values(ascending=False)[:10].index
sns.heatmap(df_cancer[top_features].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix - Top 10 Features (Breast Cancer Dataset)')
plt.tight_layout()
plt.show()

# Model Training and Analysis
scaler_log = StandardScaler()
X_log_scaled = scaler_log.fit_transform(X_log)
X_log_train, X_log_test, y_log_train, y_log_test = train_test_split(
    X_log_scaled, y_log, test_size=0.2, random_state=42
)

logistic_model = LogisticRegression(random_state=42, max_iter=2000)
logistic_model.fit(X_log_train, y_log_train)

# Feature importance for logistic regression
feature_importance_log = pd.DataFrame({
    'feature': cancer.feature_names,
    'importance': abs(logistic_model.coef_[0])
})
feature_importance_log = feature_importance_log.sort_values('importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='importance', y='feature', data=feature_importance_log.head(10))
plt.title('Top 10 Feature Importance in Logistic Regression')
plt.show()

# Cross-validation scores
cv_scores_log = cross_val_score(logistic_model, X_log_scaled, y_log, cv=5, scoring='accuracy')
print("\nCross-validation accuracy scores:", cv_scores_log)
print("Mean accuracy:", cv_scores_log.mean())
print("Standard deviation:", cv_scores_log.std())

# ROC Curve
from sklearn.metrics import roc_curve, auc
y_pred_proba = logistic_model.predict_proba(X_log_test)[:, 1]
fpr, tpr, _ = roc_curve(y_log_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
